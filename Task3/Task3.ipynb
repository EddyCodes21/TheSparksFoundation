### Data Visualisation

```python
plt.figure(figsize=(15,10))
sns.countplot(df1.State)
plt.xticks(rotation=90)
plt.show()
```

###### Maximum products are sold from California and minimum products are sold from Wyoming

```python
plt.figure(figsize=(15,10))
df1.Region.value_counts().plot.pie(autopct = '%.2f%%')
plt.show()
```

###### Maximum sales is from the West region and minimum sales is from South region 

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',color='r', data=df1)
plt.show()
```

###### The Profit Discount trend seems to be decreasing, which is also intuitive as discount increases profit margin decreases. But small discounts are good as it tends to attract more customers the increasing sales because of this balances the profit margin, this hypothesis is also supported by the above observation
###### But the sudden rise at 60% discount is probably beacuse of the fact that such high rate of discount places the product far ahead of it competitors which might indicate a monopoly situation, hence the steep rise in profit value (launch of Jio in india is a classic example of the above hypothesis)

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['r','b'],figsize=(8,5))
plt.ylabel('Profit and Sales')
plt.show()
```

###### Profit and Sales is maximum in Consumers

```python
df2 = df1.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales')
df2[:].plot.bar(color=['blue','orange'],figsize=(15,8))
plt.title('Profit / Loss and Sales across states')
plt.xlabel('States')
plt.ylabel('Profit / Loss and Sales')
plt.show()
```

###### Texas suffers the maximum loss while California has the maximum profit 

```python
plt.figure(figsize=(15,8))
plt.bar('State','Discount', color='m', data=df1)
#plt.xlabel('Quantity') 
#plt.ylabel('Profit') 
#plt.title('Quantity vs Profit ')
plt.xticks(rotation=90)
plt.show()
```

###### Texas and Illinois are offering the maximum discount

#### From the last two observations we can conclude that the business strategy in Texas is not going well. They are offering maximum discount which is leading to the maximum loss as well. Texas has an adequate amount of consumers but their business strategy is resulting in a loss. Hence, new business strategies need to be started in Texas.

#### Calfornia on the other hand is performing excellent with their business strategies.
